# ARIMA Models

**ARIMA**, which stands for **AutoRegressive Integrated Moving Average**, is a forecasting algorithm that is suited to time series data either to better understand the data or to predict future points in the series. These models should be used with stationary data. We cannot use it with random walk data, since this type of data is not stationary. 

## Stationarity in Time Series

A time series is said to be stationary if its statistical properties do not change over time. In other words, it has constant mean, variance, and covariance at different time intervals.

### Weak Stationarity
- **Definition:** A time series is weakly stationary if its mean and variance are constant over time, and the covariance between the two periods only depends on the distance or lag between the two periods and not on the actual time at which the covariance is computed.
- **Example:** A time series that fluctuates around a constant mean and has a constant variance over time, without any predictable patterns such as trends or seasonal variations.

### Strong Stationarity
- **Definition:** A time series is strongly stationary if all its moments (mean, variance, skewness, kurtosis, etc.) are invariant over time, not just the first two moments as in weak stationarity.
- **Example:** A perfectly random series with no discernible patterns or structures (e.g., white noise).

### Testing for Stationarity
- **Check ACF plot:** The ACF plot of stationary series drops to zero quickly, while that of non-stationary series tends to decrease slowly
- **Formal Tests:** *The Augmented Dickey-Fuller test* tests the null hypothesis $H_0$ that a series is non-stationary.

## Data Transformations: Differencing
In order to transform a non-stationary time series into a stationary one, we can calculate a differenced version of our original series:
- **Simple/first-order differencing:** Compute the differences between consecutive observations to eliminate trends in the series.
$y′_t = y_t − y_{t−1}$
- **Second-order differencing:** If the differenced data still exhibits trends, it may be necessary to difference the series another time.
$y′′_t = y_t − y_{t−1} − (y_{t−1} − y_{t−2}) = y_t − 2y_{t−1} + y_{t−2}$
- **Seasonal differencing:** Calculate the difference between an observation and the previous observation from the same season.
$y′_t = y_t − y_{t−m}$, where *m* is the seasonal frequency.

All of the above methods can be conmined as needed to make a series stationary.



## Data Trasnformations: Variance Stabilization

If the variation of a series increases with the level of the series, certain transformations can stabilize the variance:
- **Logarithmic transformation** (only if series does not take on values ≤ 0): $w_t = log(y_t)$, where $log()$ is the natural logarithm. Changes in log values can be interpreted as percentage changes on the original scale. You will recognize you need to do logarithmic transformations if the absolute value of the level increases as the overall value of the curve increases overtime. 
- **Box-Cox transformation** (if series also takes on values ≤ 0):$ w_t = \begin{cases} \log(y_t) & \text{if } \lambda = 0 \\\frac{(y_t^\lambda - 1)}{\lambda} & \text{otherwise} \end{cases}$ where λ is selected such that the size of the variation is similar across the whole series. (After analyzing and forecasting the Box Cox-transformed data, we need to back-transform the series to obtain forecasts on the original scale.) If λ, then we just have tha logarithmic transformation.


## Backshift Notation
Lagged values of \( y_t \) can be expressed with the **backward shift operator** \( B \):
E.g., \( B y_t = y_{t-1} \) and \( B(B y_t) = B^2 y_t = y_{t-2} \)

\( B \) can be convenient for representing differenced series:
- 1st-order differencing: $y_t' = y_t - y_{t-1} = y_t - B y_t = (1 - B)y_t$
- 2nd-order differencing:
$$
y_t'' = (y_t - y_{t-1}) - (y_{t-1} - y_{t-2}) = y_t - 2y_{t-1} + y_{t-2} = y_t - 2By_t + B^2 y_t = (1 - 2B + B^2)y_t = (1 - B)^2 y_t
$$
- \( d \)th-order differencing: \( (1 - B)^d y_t \)
- seasonal differencing followed by 1st-order differencing:
$$
(1 - B)(1 - B^m) y_t = (1 - B - B^m + B^{m+1}) y_t = y_t - y_{t-1} - (y_{t-m} - y_{t-m-1})
$$

- **What It Does**: The backshift operator *B* is a notational convenience used in time series analysis. Applying *B* to a time series shifts all the observations by one time period backward. For example:
  - $B y_t = y_{t-1}$: Applying *B*once shifts the series one period back.
  - $B^2 y_t = y_{t-2}$: Applying *B* twice shifts the series two periods back.
  
  The backshift operator makes it easier to express lags and differencing in a more compact form. For instance, $(1 - B)^d y_t$ represents the series differenced *d* times, which is a critical step in transforming a non-stationary series to a stationary one in the ARIMA modeling process.

### Examples of Using the Backshift Operator in ARIMA:
- For an ARIMA(1,1,1) model, the equation can be written using the backshift operator as:
  
  $$(1 - \phi_1 B)(1 - B)y_t = (1 + \theta_1 B)\epsilon_t$$
  
  This denotes that the series has been first differenced once (to address the *I* component), and then the AR and MA components are applied.

When these variables and operators are combined in an ARIMA model, the ultimate goal is to find the parameters $\phi$, *d*, $\theta$ that best capture the underlying process generating the time series. This is often done through methods such as maximum likelihood estimation or using criteria like AIC/BIC for model selection.

## Components of ARIMA:

### AutoRegressive
- **AR (AutoRegressive):** Refers to the use of past values in the regression equation for the time series. The number of lagged observations included in the model is termed the order of the AR model. It is denoted by \(p\). $y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \epsilon_t$
    
    There are several special cases for AR(*p*)-models. 
    - AR(1): if $\phi_1 = 0, y_t$ is **white noise**
    - AR(1): if $\phi_1 = 1$ and $c = 0, y_t$ is **random walk** 
    - AR(1): if $\phi_1 = 1$ and $c \neq 0, y_t$ is **random walk with a drift**
    
    Usually AR(*p*)-models are only used for stationary data. For an AR(*p*)-model to be stationary, certain conditions must hold:
    - AR(1): If $ - 1 < \phi_1 < 1, y_t $ is **stationary**
    - AR(2): If $ - 1 < \phi_1 < 1$ and $\phi_1 + \phi_2 < 1, y_t $ is **stationary**
    - AR(*p*) with $p > 2:$ The restrictions for stationarity are relatively complicated. 

- **Variables**:
  - $y_t$: The value of the time series at time $t$.
  - $\phi_i$: The coefficient of the lag *i* value of the time series.
  - $c$: The constant term or intercept in the model (if applicable).
  - $\epsilon_t$: The error term at time *t*, which is assumed to be white noise.
  - *p*: The order of the autoregressive model, indicating the number of lags to be included.
- **What It Does**: The AR component models the current value of the time series as a linear combination of its past values plus an error term. The number *p* indicates how many past values are considered. The AR component captures the relationship between an observation and its predecessors.


### Integrated
- **I (Integrated):** Represents the differencing of raw observations to allow for the time series to become stationary, i.e., data values are replaced by the difference between the data values and the previous values. The number of times the raw observations are differenced is denoted by *d*.

- **Variable**:
  - *d*: The degree of differencing.
  
- **What It Does**: The integration part involves differencing the time series a certain number of times (denoted by *d*) to make it stationary, i.e., to stabilize the mean of the time series if it shows trends or seasonality.



### Moving Average
- **MA (Moving Average):** Incorporates the dependency between an observation and a residual error from a moving average model applied to lagged observations. The number of lagged forecast errors in the prediction equation is termed the order of the MA model, denoted by \(q\).
    - **Approach**: The outcome variable is modeled as a weighted average of past forecasting errors. 
    - A moving average model of order *q* (MA(*q*)-model) is defined as the followeing, where $\epsilon_t$ is the white noise: 
  $$y_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q}$$
    - **Careful!** The MA(*q*) model (= weighted average of past values of εt for prediction purposes) should not be confused with the moving average approach used for smoothing time series and for dealing with missing values (= weighted average of $y_t$ in a moving window for smoothing purposes).
    - **Characteristics**: The parameters $\theta_1$, ..., $\theta_p$ are selected so that the model captures the shape of the series, the variance of εt to capture the scale.
    
- **Variables**:
  - $\theta_i$: The coefficient of the lagged forecast error at lag *i*.
  - *q*: The order of the moving average model, which specifies the number of lagged forecast errors included.
- **What It Does**: The MA component models the time series as a linear combination of the past forecast errors. The number *q* indicates how many past forecast errors are used. It aims to capture the shock from an unexpected event that affects the observation.

    
### ARMA (*q, p*) Process
- The ARMA(*p, q*) process combines AR(*p*) and MA(*q*) to capture both autoregressive relationships and relationships in error terms: $$y_t = \phi_1 y_{t−1} + ... + \phi_p y_{t−p} + ε_t + θ_1 ε_{t−1} + ... + θ_q ε_{t−q}$$
- E.g.: An ARMA(2,1) process is given as: $$ y_t = \phi_1 y_{t−1} + \phi_2 y_{t−2} + ε_t + θ_1 ε_{t−1} $$

- **What It Does**: It combines both AR and MA components. An ARMA model captures the autocorrelation and the moving average aspects of the time series.


### ARIMA (*p, d, q*) Process
Modelling a time series with a non.seasonal ARIMA (*p, d, q*) model
- **Example**: An ARIMA (2, 1, 1) process is given as: $$y_t - y_{t-1} = \phi_1 (y_{t-1} - y_{t-2}) + \phi_2 (y_{t-2} - y_{t-3}) + \epsilon_t + \theta_1 \epsilon_{t-1}$$
- Or, in **backshift notation**: $$(1 - \phi_1 B - \phi_2 B^2) * (1 - B) * y_t = (1 + \theta_1 B) \epsilon_t$$

- **What It Does**: The ARIMA model extends ARMA by including the integration component, making it suitable for analyzing non-stationary time series that become stationary after differencing.

### Seasonal ARIMA $(P, D, Q)_m$
- *D*-degree-seasonal differencing helps to remove seasonality and make the data stationary; *P* and *Q* help account for relationships between the same season in different seasonal cycles. 
- The seasonal ARIMA-models are best written with backshift notation as they else quickly become complicated. 
- **Example**: For monthly data, a potentially suitable seasonal ARIMA-models is ARIMA$(0, 1, 1)(0, 0, 1)_{12}$:
    $$(1 - B) * y_t = (1 + \theta_1 B) * (1 + \Phi_1 B^12) \epsilon_t$$
    $$y_t - y_{t-1} = \epsilon_t + \theta_1 \epsilon_{t-1} + \Phi_1 \epsilon_{t-12} + \theta_1 \Phi_1 \epsilon_{t-13}$$
    
    or ARIMA$(1, 1, 1)(1, 1, 1)_{12}$:
    $$(1 - \phi_1 B)(1-\phi_1 B^{12})(1 - B)(1 - B^{12}) y_t = (1 + \theta_1 B)(1 + \Phi_1 B^{12}) \epsilon_t$$

The seasonal ARIMA model, denoted as ARIMA$(P, D, Q)_m$, extends the ARIMA model to include seasonal elements. The components are defined as:

- $P$: The number of seasonal autoregressive (SAR) terms.
- $D$: The number of seasonal differences required to make the series stationary.
- $Q$: The number of seasonal moving average (SMA) terms.
- $m$: The number of periods in each season.

In this notation:
- Seasonal Differencing ($D$) refers to the process of computing the difference between an observation and its counterpart in the previous cycle to remove seasonality.
- Seasonal Autoregressive Terms ($P$) capture the relationship of an observation with its seasonal lags.
- Seasonal Moving Average Terms ($Q$) are used to model the residual errors associated with the seasonal lags.

The backshift operator $B$ is used to denote lags in the model, where $B^k y_t$ refers to the value of the series at time $t-k$.

For example, ARIMA$(0, 1, 1)(0, 0, 1)_{12}$ suggests a seasonal model for monthly data without autoregressive terms, a single non-seasonal difference, and one seasonal moving average term. The equation can be written as:
$$(1 - B)y_t = (1 + \theta_1 B)(1 + \Phi_1 B^{12})\epsilon_t$$
expanding to:
$$y_t - y_{t-1} = \epsilon_t + \theta_1 \epsilon_{t-1} + \Phi_1 \epsilon_{t-12} + \theta_1 \Phi_1 \epsilon_{t-13}$$

Similarly, ARIMA$(1, 1, 1)(1, 1, 1)_{12}$ includes both non-seasonal and seasonal AR and MA terms, along with differencing, to accommodate complex seasonality patterns. The model equation is:
$$(1 - \phi_1 B)(1-\phi_1 B^{12})(1 - B)(1 - B^{12})y_t = (1 + \theta_1 B)(1 + \Phi_1 B^{12})\epsilon_t$$

### Forecasting with an ARIMA Model
**Point estimates:**
- Bring yt to the left-hand side of the ARIMA equation.
- Replace *t* with $T + h$.
- Replace future observations with their forecasts, future errors with zero and past errors with the corresponding residuals
**Prediction intervals:**
- The calculation of prediction intervals for ARIMA-models is based on the assumption that the residuals are uncorrelated and normally distributed.
- Check if this holds (via histogram and residual ACF plot), before calculating prediction intervals.
- The prediction interval for 1-step-ahead forecasts can be calculated as $\hat{y}_{T+1|T} \pm 1.96 \hat{\sigma}_e$ with $\hat{\sigma}_e$ the standard deviation of the residuals.
- For most ARIMA models, the calculation of multi-step forecast prediction intervals is relatively complex but they are provided in Python

#### ARIMA(1,1,1) forecasting example
**Example:** Forecasting based on an estimated ARIMA(1,1,1) model:

- The original ARIMA(1,1,1) model equation
$$
(1 - \phi_1 B)(1 - B) y_t = (1 + \theta_1 B) \varepsilon_t
$$

- The model expressed as a first-order difference equation
$$
\Longleftrightarrow (y_t - y_{t-1}) = \phi_1 (y_{t-1} - y_{t-2}) + \varepsilon_t + \theta_1 \varepsilon_{t-1}
$$

- Rewriting the above equation in terms of y_t
$$
\Longleftrightarrow y_t = y_{t-1} + \phi_1 y_{t-1} - \phi_1 y_{t-2} + \varepsilon_t + \theta_1 \varepsilon_{t-1}
$$

- Explanation for a 1-step-ahead forecast
For a 1-step-ahead forecast, replace *t* with $T + 1$:

- The equation for a 1-step-ahead forecast
$$
y_{T+1} = y_T + \phi_1 y_T - \phi_1 y_{T-1} + \varepsilon_{T+1} + \theta_1 \varepsilon_T
$$

- Assuming the expectation of future error is zero
$$
\text{where } \mathbb{E}(\varepsilon_{T+1}) = 0 \text{ and } \varepsilon_T \text{ is estimable by } r_T
$$

- The final 1-step-ahead forecast equation
$$
\therefore \hat{y}_{T+1|T} = y_T + \phi_1 y_T - \phi_1 y_{T-1} + \theta_1 r_T
$$


For a multi-step forecast:

- The equation for a multi-step forecast, h > 2
$$
\hat{y}_{T+2|T} = \hat{y}_{T+1|T} + \phi_1 \hat{y}_{T+1|T} - \phi_1 y_T
$$

$$
\hat{y}_{T+h|T} = \hat{y}_{T+h-1|T} + \phi_1 \hat{y}_{T+h-1|T} - \phi_1 \hat{y}_{T+h-2} \text{ for } h > 2
$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error,mean_squared_error
import pmdarima as pm

import warnings
warnings.filterwarnings('ignore')
```

```python
#### Manually make airpassenger series stationary
airline_passengers = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
airline_passengers.plot(figsize=(20, 8));
```

```python
# Check if (obviously not stationary) airpassenger series is stationary with Augmented-Dickey-Fuller test 
adresults = adfuller(airline_passengers['Passengers'])
print('ADF Statistic: %f' % adresults[0])
print('p-value: %f' % adresults[1])
print('Critical Values:')
for key, value in adresults[4].items():
     print('\t%s: %.3f' % (key, value))
```

```python
# Check for stationarity with ACF
plot_acf(airline_passengers['Passengers']);
```

```python
# Logarithmize series to stabilize variation
airline_passengers['LogPassengers'] = np.log(airline_passengers['Passengers'])
airline_passengers['LogPassengers'].plot(figsize=(20, 8));
```

```python
# Calculate 1st difference to remove trend
airline_passengers['1stdiff'] = airline_passengers['LogPassengers'].diff(1)
airline_passengers['1stdiff'].plot(figsize=(20, 8));
```

```python
# Calculate seasonal difference to remove seasonality
airline_passengers['seasonaldiff'] = np.nan
for i in range(12, len(airline_passengers)):
    airline_passengers['seasonaldiff'][i] = airline_passengers['1stdiff'][i] - airline_passengers['1stdiff'][i - 12]
airline_passengers['seasonaldiff'].plot(figsize=(20, 8));
```

Here, this data might be stationary, but there are issues with the variance. We could work with it, but we can try to use the Augmented test to see. 

```python
# Check if series is stationary with Augmented Dickey Fuller test
adresults = adfuller(airline_passengers['seasonaldiff'].dropna())
print('ADF Statistic: %f' % adresults [0])
print('p-value: %f' % adresults [1])
print('Critical Values:')
for key, value in adresults [4].items():
     print('\t%s: %.3f' % (key, value))
```

We can reject the null hypothesis, meaning as can make the assumtion that it is in fact stationary. 

```python
#Check for stationarity with ACF plot
plot_acf(airline_passengers['seasonaldiff'].dropna());
```

```python
plot_pacf(airline_passengers['seasonaldiff'].dropna());
```

```python
# Manually fit ARIMA to data on change in consumption in the US
US_change = pd.read_csv('US_change.csv', index_col='Quarter', parse_dates=True)
US_change['Consumption'].plot(figsize=(20, 8));
```

```python
# Check for stationarity with ADF
adresults = adfuller(US_change['Consumption'])
print('ADF Statistic: %f' % adresults [0])
print('p-value: %f' % adresults [1])
print('Critical Values:')
for key, value in adresults [4].items():
     print('\t%s: %.3f' % (key, value))
```

```python
# Check ACF and PACF to determine p and q
plot_acf(US_change['Consumption']);
```

```python
plot_pacf(US_change['Consumption']);
```

In the PACF was the case, we could fit an ARIMA model with a 3, 0, 0. 

```python
# Create training and test set
train = US_change['Consumption'].iloc[:-25]
test = US_change['Consumption'].iloc[-25:]

# Fit ARIMA(3,0,0) model
arima_model = ARIMA(train, order=(3,0,0))

# Predict series for test period
arima_prediction = arima_model.fit().predict(start=test.index[0], end=test.index[-1])

# Plot training and test set and prediciton
train.plot()
test.plot()
arima_prediction.plot(figsize=(20, 8));
```

```python
# Fitting the auto_arima without seasonality results in error "It is likely your data is non-stationary." 
# -> Insert the only possible seasonality in this time series m = 4
auto_arima_model = pm.auto_arima(train,trace=True,suppress_warnings=True, seasonal = True, m=4)
# Get info on fitted model
auto_arima_model.summary()
```

```python
# Predict values for test period and get prediction interval
auto_arima_prediction, confint = auto_arima_model.predict(n_periods=len(test), return_conf_int=True)

# plot training and test data as well as forecast and prediction interval
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(train)
ax.plot(test)
ax.plot(test.index, auto_arima_prediction)
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.2)
```

```python
# Compare prediction accuracy of manually chosen and automatically identified model specification 
print('Manually chosen model: Mean abs. error %f' % mean_absolute_error(arima_prediction,test))
print('Manually chosen model: Mean squ. error %f' % mean_squared_error(arima_prediction,test))

print('Automatically chosen model: Mean abs. error %f' % mean_absolute_error(auto_arima_prediction,test))
print('Automatically chosen model: Mean squ. error %f' % mean_squared_error(auto_arima_prediction,test))
```

```python
# Automatically find optimal seasonal ARIMA for Log-Passenger series
train = airline_passengers['LogPassengers'].iloc[:120]
test = airline_passengers['LogPassengers'].iloc[120:]

sarima_model = pm.auto_arima(train,trace=True,suppress_warnings=True,seasonal=True, m=12)
sarima_model.summary()
```

```python
# get prediction and confidence interval
sarima_prediction, confint = sarima_model.predict(n_periods=len(test), return_conf_int=True)

# plot training and test data as well as forecast and prediction interval
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(train)
ax.plot(test)
ax.plot(test.index, sarima_prediction)
ax.fill_between(test.index, confint[:,0], confint[:,1], color='red', alpha=0.2)
```

```python
# Goodness-of-fit
# Fit of fitted series to training data
train_pred = sarima_model.predict_in_sample(start=0, end=-1)

fig, ax = plt.subplots()
ax.plot(train[12:])
ax.plot(train.index[12:],train_pred[12:]);
```
