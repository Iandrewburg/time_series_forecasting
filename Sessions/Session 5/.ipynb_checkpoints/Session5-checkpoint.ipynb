## Linear Regression Models
---

### Simple Linear Regression Model

$y_t = \beta_0 + \beta_1 x_t + \epsilon_t$
- $\epsilon_t$ is the error term, i.e., the deviation from the linear model in *t*
- The coefficients can be interpreted just as in simple linear regression with cross-sectional data: An increase in $x_t$ by one unit leads to a change in $y_t$ by $\beta_1$ units. 
- The variables $y_t$ and/or $x_t$ can be a non-linear transformation (e.g. logarithmic) or the original variables, if the relatioship between the transformed series is linear, while that of the original series is not. 

### Multiple Linear Regression Model

$y_t = \beta_0 + \beta_1 x_{1t} + \beta_2 x_{2t} + ... + \beta_k x_{kt} + \epsilon_t$

- $\epsilon_t$ is the error term, i.e., the deviation from the linear model in *t*
- The coefficients can be interpreted just as in multiple linear regression with cross-sectional data: The coefficients $β_1, ..., β_k$ measure how $y_t$ changes in response to a change in one predictor after taking into account the influence of the other predictors.
- The variables $y_t$ and/or $x_{1t}$, ..., $x_{kt}$ can be a non-linear transformation (e.g. logarithmic) of the original variables if the relationship between the transformed series is linear, while that of the original series is not.

### Linear Regression Estimation

Estimation based on Least Squares, i.e., by choosing $\beta_0, \ldots, \beta_k$ that minimize the squared error:

$$
\sum_{t=1}^{T} e_t^2 = \sum_{t=1}^{T} (y_t - \beta_0 - \beta_1x_{1,t} - \beta_2x_{2,t} - \ldots - \beta_kx_{k,t})^2
$$

Like in regressions with cross-sectional data, common measures for goodness-of-fit can be employed like e.g.:

$$
R^2 = \frac{\sum_{t=1}^{T} (\hat{y}_t - \bar{y})^2}{\sum_{t=1}^{T} (y_t - \bar{y})^2}, \quad (0 \leq R^2 \leq 1)
$$

measuring the proportion of variation in $y_t$ that is captured by the model.

$\Rightarrow R^2$ should be calculated in test data to avoid overfitting.

### Linear Model Evaluation
**Model evaluation**
Certain checks should be run to evaluate if the model assumptions hold (again similar to linear regression with cross-sectional data), namely:
- Create a histogram of the residuals $et = y_t - \hat{y}_t$ : Are the $e_t$ approximately normally distributed?
- Plot the $e_t$ against predictors: If there is a pattern in any plot, the relationship between $y_t$ and the respective predictor may be non-linear.

Additionally, certain checks particular to time series regression should be employed.
Check for **autocorrelation in the residuals**:
- Plot the residual ACF.
- Run formal tests to test for autocorrelation.
⇒ Autocorrelation in the residuals indicates that the model is inefficient as it doesn’t account for all information available. The model is unbiased but the prediction intervals are unnecessarily wide.


Check for **time-dependent heteroscedasticity**:
- Create a residual time plot: Is the residual variance changing over time?
⇒ In case of heteroscedasticity, the model is unbiased but the prediction intervals may be wrong.

### Spurious regression
- Spurious relations are apparent connections/correlations between variables (or time series) that are actually coincidental and not indicative of any true relationship.
- In time series regression, there is a particularly high risk of modeling spurious relationships when dealing with non-stationary, and especially with trending, series because technically all series with a similar trend are highly correlated whether there is an underlying theoretical relationship or not.

⇒ Important to consider whether a relationship between yt and potential predictors is theoretically plausible. Additionally, high R2 combined with high autocorrelation in residuals can be an indicator of spurious relations.


### Potential Predictors
- A time series $z_t$ that is correlated with $y_t$ , e.g. to model ice-cream sales $y_t$ one can include weather $z_t$ as predictor
- Lagged values of a series, e.g. if $z_t$ measures the advertisement expenditures, $x_{i,t} = z_{t−1}, x_{i+1,t} = z_{t−2}, ...$ can be included in the model to capture the *longer-term effect* of advertisement
- A variable $x_{i,t} = t$ with $t = 1, ..., T$ can be included to capture a **linear trend** in the series
- A series of **seasonal dummies**, e.g. for a quarterly series:$d_{Q1,t} , d_{Q2,t} , d_{Q3,t} , d_{Q4,t}$ , where $d_{Q1,t} = 1$ if t falls into the first quarter and 0 otherwise, $d_{Q2,t} = 1$ if *t* falls in the first quarter and 0 otherwise,...

### Fourier Terms as Potential Predictors
Fourier series are a tool for expressing an arbitrary periodic function (like the seasonality pattern of a time series) as a weighted sum of sine and cosine terms. 

- The Fourier terms are given by:
$$
x_{1,t} = \sin\left(\frac{2\pi t}{m}\right), \quad x_{2,t} = \cos\left(\frac{2\pi t}{m}\right), \quad x_{3,t} = \sin\left(\frac{4\pi t}{m}\right),
$$
$$
x_{4,t} = \cos\left(\frac{4\pi t}{m}\right), \quad x_{5,t} = \sin\left(\frac{6\pi t}{m}\right), \quad x_{6,t} = \cos\left(\frac{6\pi t}{m}\right), \ldots
$$

- *How to include Fourier terms in the model*: The terms can be included in pairs of one sine and one cosine term, i.e., either the first two, the first four, ..., up to the first *m*.

- *Benefits of using Fourier terms*: Usually a few Fourier terms are sufficient to capture the seasonality of a time series, making them particularly advantageous for series with large *m*.


### Selecting Predictors
Once all potential predictors are identified, we must identify the set of predictors that yields the highest prediction accuracy when included in the model.

There are two commonly used approaches to screen through all possible sets of predictors:
- **Best subset regression**: compare the prediction accuracy of all possible models ⇒ if computationally not to challenging, this approach should be preferred.
- **Backwards step-wise regression**: Start with a model containing all potential predictors, remove one predictor at a time and keep the model with the highest predictive accuracy.

#### Measures to Compare Candidate Models with Different Sets of Predictors
- The adjusted $R^2$ denoted as $\bar{R}^2$, is a modified version of $R^2$ that penalizes for the number of predictors *k*, i.e. it accounts for the model’s complexity and prevents overfitting: 
$$ \bar{R}^2 = 1 - (1 - R^2) \frac{T - 1}{T - k - 1} $$
- The information criteria AIC, AICc and BIC can be used as discussed in Session 2.
- **Cross validation** as discussed in Session 2.

### Forecasting with Linear Regression
To forecast $y_{T + h}$ with linear time series regression models, we have to plug the values for $x_{1,T + h}, ...., x_{k,T + h}$ into the estimated model. 
- For $x_{i,t}$ that capture deterministic features like calendrical specificities (number of business days, day-of-the-week dummies) or seasonality (Fourier terms or seasonality dummies), these values are known for any *T + h*.
- For $x_{i,t}$ that measure lagged values of a variable, future values are known as long as the h is not larger than the lag.
- For $x_{i,t}$ that measure other variables, we can simply set $x_{i,T + H}$ to a likely values, run through different scenarios, use available forecasts (for weather, economic growth,...) or predict the predictors using any suitable time series forecasting approach.

### Forecasting with Linear Regression Prediction Intervals

*Explaining what the prediction intervals depend on*

The width of the prediction intervals for the forecast $\hat{y}_{T+h}$ depends on the residual variance and on how far away the predictors are from their average. For example, for a simple linear regression model, the 95%-prediction interval for $\hat{y}_{T+h}$ is given by:

*The prediction interval formula*

$$
\hat{y}_{T+h} \pm 1.96\hat{\sigma}_e\sqrt{1 + \frac{1}{T} + \frac{(x_{T+h} - \bar{x})^2}{(T - 1)s_x^2}}
$$

*Explanation of the terms in the formula* 

where $\hat{\sigma}_e = \sqrt{\frac{1}{T-k-1} \sum_{t=1}^{T} e_t^2}$ is the standard error of the regression and $s_x^2$ is the standard deviation of $x$. Be careful: the prediction interval does not take into account possible uncertainty in predicting the predictors.



## Dynamic Regression Models

- Dynamic regression models extent ARIMA models by including predictors xt , or, put differently, dynamic regression models model the potential autoregression in the errors of a regression model.
- E.g.: A regression model where the error term ηt follows an ARIMA(1,1,1) model is given by: $$y_t = β_0 + β_1 x_{1,t} + ... + β_k x_{k,t} + η_t$$ $$(1 − φ_1 B)(1 − B) η_t = (1 + θ_1 B) ε_t$$ , where $ε_t$ is white noise.
- If the above model was the true model underlying a time series, simply estimating the regression model while ignoring the autoregressive error term would result in ..
    
    ... the model not being the best option given the available information and
    
    ... in incorrect prediction intervals.
    
---

### Estimating Dynamic Regression Models
- To estimate a regression model with ARIMA errors, $y_t$ and all $x_{1,t} , ..., x_{k,t}$ must first be made stationary 
    
    ⇒ Apply the same transformations to all series to maintain the relationship between variables and ensure interpretability of the results.
- To determine the optimal set of predictors, compare the $AIC_c$ of all candidate models.

### Foresting Dynamic Regression Models
- Forecast the regression part and the ARMA part of the model separately and combine it.
- If the predictors are unknown in $T + h$, they may be forecasted before being plugged into the regression model or set to specific values (Careful: the prediction intervals do not take into account the uncertainty in predicting the predictors)

### Advantages and Disadvantages of Dynamic Regression Models
- Advantage: Dynamic regression models allow us to account for autocorrelation in the data and at the same time take advantage of relationships between $y_t$ and other series $x_t$. 
- Advantages: They are very flexible in capturing seasonality, calendar peculiarities and events
- Disadvantage: Dynamic regression models only capture linear relationships

## State Space Models
---

### Underlying Idea of State Space Models

- There is an unobservable system that determines the development of the observed time series due to internal dynamics and external forces that affect the system.
- In every t, the system is in a certain state but we cannot observe the true state directly - we observe the state with some noise added
- The aim of state space modeling is to model the dynamic system producing the data we observe rather than the noisy data itself.
    - State space models are particularly suitable when we have strong theories/knowledge about how the underlying system works.
- The model coefficients can change over time as the model is constantly updated based on the most recent data points.


### Hidden Markov Model
- Assumption about the underlying process: 
    - The states we observe are generated by a **Markov process that jumps between discrete states**
    - Markov processes are memory-less, i.e., the state of the process in $t + 1$ depends only on the state in $t$. 
    - The dynamics within a Markov process can be fully described based on the transition probabilities between any *t* and $t + 1$, which are usually represented in matrix form. 
    
- Assumption about the observed series: We cannot observe which state the system is in but the observations give clues about it. 
- Potential Applications: Modeling an economic time series to capture underlying economic states and transitions, or financial series to identify regime shifts in the financial market. 

#### Fitting the Hidden Markov Model
Fitting a Hidden Markov Model to a time series is an unsupervised learning task (we have to identify the hidden states without knowing any of them a priori).

**Procedure**:
Procedure:
1 Make assumptions about the number of underlying states and the
relationship between the true states q1, ...qT and the observed series
y1, ...yT .
E.g.: 2 states produce a normally distributed output (the values we
observe, yt ). Each state qi is characterized by mean μi and standard
deviation σi. The probability to change from one state to the other is p.
2 Set the parameters that characterize the states randomly (i.e., μi, σi, p in
the example).
3 For each t, determine which state is most likely to have produced yt .
4 Update the parameters based on the identified series of states.
5 Repeat steps 2 and 3 until convergence is achieved.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from tbats import TBATS
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error,mean_squared_error
import pmdarima as pm
from prophet import Prophet
from sktime.transformations.series.fourier import FourierFeatures

from statsmodels.graphics.tsaplots import plot_acf


from pmdarima import pipeline
from pmdarima import model_selection
from pmdarima import preprocessing as ppc
from pmdarima import arima
```

```python
######### Overlapping seasonal patterns. 
#### TBATS
# Here: Sales data with yearly and weekly seasonality
# Sales data from different stores. One time series per store and article, all stored under each other 
sales = pd.read_csv('sales.csv', index_col='date', parse_dates=True)
# time series of interest: sales of item 1 in store 1
sales_s1i1 = sales[(sales['store'] == 1) & (sales['item'] == 1)] 
# create training and test set
train = sales_s1i1['sales'].iloc[:(len(sales_s1i1)-365)]
test = sales_s1i1['sales'].iloc[(len(sales_s1i1)-365):] 


# Plot training and test data
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(train)
ax.plot(test.index, test);
```

```python
# Fit TBATS model and predict following year
TBATSmodel = TBATS(seasonal_periods=(7, 365.25)).fit(train)
TBATSforecast = TBATSmodel.forecast(steps=365)
```

```python
# Plot training data and forecast
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(train)
ax.plot(test.index, TBATSforecast, color = "green");
```

```python
###### Fourier
# Create Fourier terms  and include them in ARIMA model
pipe = pipeline.Pipeline([
    ("fourier", ppc.FourierFeaturizer(m=365.25, k=4)),
    ("arima", arima.AutoARIMA(seasonal=True, m = 7, D=1,suppress_warnings=True, trace = True))])

# Data must be inserted as list
pipe.fit(list(train))
```

```python
# Estimate forecasts and prediction intervals for 1 year
Fourierprediction, conf_int = pipe.predict(n_periods=365, return_conf_int=True)
```

```python
# Plot training data and forecast
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(train)
ax.plot(test.index, Fourierprediction, color = "green");
```

```python
######## Dynamic regression model
# Time series of interest: Quarterly change in consumption in the US, 
US_change = pd.read_csv('US_change.csv', index_col='Quarter', parse_dates=True)
# Create training and test set
train = US_change[:round(len(US_change)*0.8)]
test = US_change[round(len(US_change)*0.8):]
# Plot training and test data
train['Consumption'].plot(figsize=(20, 8))
test['Consumption'].plot(figsize=(20, 8));
```

```python
# Fit the model with change in income as predictor variable
dynreg_fit = pm.auto_arima(train.Consumption, train[['Income']])
dynreg_fit.summary()
```

```python
# Residual diagnostics
# Calculate residuals
train["Residuals"] = train["Consumption"] - dynreg_fit.arima_res_.fittedvalues
# Plot ACF plot for residuals
plot_acf(train["Residuals"]);
```

```python
# Plot the residuals
train["Residuals"].plot(figsize = (20,8));
```

```python
# Set the income change for the prediction period to its mean in the past
exog = np.repeat(train.Income.mean(), len(test))
# 
dynreg_prediction = dynreg_fit.arima_res_.get_forecast(len(exog), exog=exog)

# plot training and test data as well as prediciton and prediction interval
fig, ax = plt.subplots(figsize = (20,8))
ax.plot(train.Consumption)
ax.plot(test.Consumption)
ax.plot(dynreg_prediction.predicted_mean)
ax.fill_between(test.index, dynreg_prediction.conf_int()["lower y"], dynreg_prediction.conf_int()["upper y"], color='red', alpha=0.2)
```

```python
###### Prophet Model
# Data from a store 
rossmann = pd.read_csv('rossmann.csv', index_col='Date', parse_dates=True)
rossmann[rossmann['Store'] == 1]['Sales'].plot(figsize=(15, 8));
```

```python
store1 = rossmann[rossmann['Store'] == 1].copy()
# concentrate on days when store is open
store1[store1['Open'] > 0]['Sales'].plot(figsize=(15, 8));
```

```python
store1_prophet = store1[store1['Open'] > 0][['Sales']].copy()
# write index in regular column
store1_prophet['Date'] = store1_prophet.index
store1_prophet.head()
```

```python
# Data in wrong order -> sort by index
store1_prophet.sort_index(inplace=True)
store1_prophet.columns = ['y', 'ds']
prophet_model = Prophet()
prophet_model.fit(store1_prophet)
```

```python
# make dataframe with dates for forecasting period
dates_forecast = prophet_model.make_future_dataframe(periods=365)
print(dates_forecast)
```

```python
# make predicitons with Prophet model
prophet_forecast = prophet_model.predict(dates_forecast)
prophet_forecast.head()
```

```python
# Plot training data and prediction 
fig, ax = plt.subplots(figsize =(15, 8))
ax.plot(store1_prophet['y'])
ax.plot(dates_forecast.ds[-365:],prophet_forecast.yhat[-365:], color = "green")
ax.fill_between(dates_forecast.ds[-365:],prophet_forecast.yhat_lower[-365:], prophet_forecast.yhat_upper[-365:], color='red', alpha=0.2)
```

```python
# Forecasting plot as provided by Prophet
prophet_model.plot(prophet_forecast);
```

```python
# Plots of trend and seasonality
prophet_model.plot_components(prophet_forecast);
```

```python
# Prphet with built-in holidays
prophet_model2 = Prophet()
# Add holidays
prophet_model2.add_country_holidays(country_name='US')
prophet_model2.fit(store1_prophet)
prediction_dates2 = prophet_model2.make_future_dataframe(periods=365)
# Forecast
prophet_forecast2 = prophet_model2.predict(prediction_dates2)
# Plot
prophet_model2.plot(prophet_forecast2);
```

```python
# Plot components
prophet_model2.plot_components(prophet_forecast2);
```
